```python
from textSummarizer.models import ModelFactory

summarizer = ModelFactory.create("extractive")
text = (
    "Artificial intelligence is reshaping healthcare, finance, and transportation. "
    "Machine learning models detect diseases from medical images."
)
summary = summarizer.summarize(text, max_length=64, strategy="map_reduce")
print(summary)
```

```python
from textSummarizer.grading.llm_judge import LLMJudge
from textSummarizer.grading.rubric import GradingRubric

judge = LLMJudge(use_llm=False)
score = judge.grade(text, summary)
rubric = GradingRubric()
print(score.to_dict())
print("Passes:", rubric.passes(score))
```

```python
from textSummarizer.pipelines.citations import summarize_with_citations

grounded = summarize_with_citations(text, summary)
print(grounded)
```
